#polars is used for data loading and manipulation.
#elfen is used for linguistic feature extraction

````python
#Step 1: Import Required Libraries
import polars as pl
import elfen
````

#Load the text dataset into a DataFrame.
#Verify the dataset structure before analysis.

````python
# Step 2: Load the Dataset
df = pl.read_csv("Random_English_Sentences.csv")

print(df.head())
print(df.shape)
````

Creates an elfen extraction pipeline.
Specifies:
Dataset
Language
Text column

````python
#Step 3: Initialize the Extractor
extractor = elfen.Extractor( 
    data=df, 
    language="en", 
    text_column="text" 
    ) 
print("Extractor created successfully!")
````

Extracts:

Type-Token Ratio (TTR)

Formula:
TTR = Unique Words / Total Words

Interpretation:
Higher TTR → Rich vocabulary
Lower TTR → Repetitive vocabulary

````python
#Step 4: Extract a Single Feature
extractor.extract("ttr")
print(extractor.data.head())
````

Features Generated:
Flesch Reading Ease
Flesch-Kincaid Grade
SMOG
Gunning Fog
ARI
CLI
LIX
RIX

Gunning Fog = 8.45 → Approximately 8th-grade reading level.
SMOG = 8.84 → Around 9 years of education needed.
Flesch Reading Ease = 70.2 → Fairly easy to read.
Flesch-Kincaid Grade = 8.1 → Suitable for middle/high school students.

````python
#Step 6: Extract Readability Features
extractor.extract_feature_group("readability")
#Step 7: Inspect Readability Results
print(
    extractor.data.select([
        "text",
        "gunning_fog",
        "smog",
        "flesch_reading_ease",
        "flesch_kincaid_grade"
    ]).head()
)
````

Surface Features:

Number of tokens
Number of sentences
Average word length

Dependency Features:

Grammatical relationships

Morphological Features:

Word structure information

````python
#Step 8: Extract Additional Linguistic Dimensions
extractor.extract_feature_group("surface")
extractor.extract_feature_group("dependency")
extractor.extract_feature_group("morphology")

print(extractor.data.head())
````

````python
#Step 9: Extract All Features
extractor.extract_features()
print(extractor.data.head())
````

elfen successfully extracted several feature groups, including:

Surface Features
Readability Features
Lexical Richness Features
Morphological Features
Dependency Features
Syntactic Features

However, when the framework attempted to extract emotion-based features, it required an external linguistic resource called the NRC Emotion Lexicon.

Error Encountered
FileNotFoundError:
NRC-Emotion-Lexicon-Wordlevel-v0.92.txt

Why This Happened

The NRC Emotion Lexicon is not distributed automatically with the elfen package due to licensing and website restrictions.

As a result, the file was not available in the expected resource directory, causing the extraction process to terminate.

Impact
Previously extracted features remained available.
Emotion and sentiment-related features could not be generated.
Full extraction of all 1,061 features was not completed.

Solution

To enable complete feature extraction:

Download the NRC Emotion Lexicon manually.
Place the file in the resource directory expected by elfen.
Re-run:
extractor.extract_features()

The extraction process successfully generated multiple linguistic feature groups. However, the execution stopped when elfen attempted to compute emotion-related features that depend on the NRC Emotion Lexicon. Since this external resource was not installed, a FileNotFoundError occurred. This highlights a common challenge in NLP research, where tools often rely on additional linguistic resources beyond the main software package.
